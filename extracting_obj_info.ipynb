```python
def getList(list_xml):
    parsed_list=[]
    for item in list_xml:
        root = et.parse(item).getroot()
        filename = root.find('filename').text
        label = root.find('folder').text
        size = root.find('size')
        bndbox = root.find('object').find('bndbox')
        width = size.find('width').text
        height = size.find('height').text
        xmin = bndbox.find('xmin').text
        ymin = bndbox.find('ymin').text
        xmax = bndbox.find('xmax').text
        ymax = bndbox.find('ymax').text
        parsed_list.append([filename,width,height, label, xmin,xmax,ymin,ymax])
    return parsed_list
```

```python
list_xml = getList(list_xml)
```

```python
df = pd.DataFrame(list_xml,columns=['filename','width','height','label','xmin','xmax','ymin','ymax'])
```

```python
def calCenter(df):
    df['center_x'] = ((df['xmin']+df['xmax'])/2)/df['width']
    df['center_y'] = ((df['ymin']+df['ymax'])/2)/df['height'] 
    df['boxWidth'] = (df['xmax']-df['xmin'])/df['width']
    df['boxHeight'] = (df['ymax']-df['ymin'])/df['height']
```

```python
df['width'] = df['width'].astype(int)
df['height'] = df['height'].astype(int)
df['xmin'] = df['xmin'].astype(int)
df['xmax'] = df['xmax'].astype(int)
df['ymin'] = df['ymin'].astype(int)
df['ymax'] = df['ymax'].astype(int)
```

```python
def path_list(xml_c_list):
    paths=[]
    for item in xml_c_list:
        root = et.parse(item).getroot()
        filename = root.find('filename').text
        path = root.find('path').text
        path = os.path.dirname(path)
        path = os.path.normpath(path)
        paths.append([path,filename])
    return paths
```

```python
paths= path_list(xml_c_list)
```

```python
cols = ['filename', 'id', 'center_x', 'center_y', 'boxWidth', 'boxHeight']
groupby_obj_train = train_df[cols].groupby('filename')
groupby_obj_test = test_df[cols].groupby('filename')
```

```python
import os
from shutil import move

def save_data(filename, folderPath, group_obj):
    # Construct source and destination file paths
    src = os.path.join('xml', filename)
    dst = os.path.join(folderPath, filename)

    # Move the file
    try:
        move(src, dst)
        # print(f'Moved: {filename}')
    except FileNotFoundError:
        # print(f'File not found: {src}')
    except Exception as e:
        # print(f'Error moving file {filename}: {e}')
        return

    # Prepare the text filename for the CSV export
    text_filename = os.path.join(folderPath, os.path.splitext(filename)[0] + ".txt")

    # Export the grouped data to a text file
    try:
        group_data = group_obj.get_group(filename)
        group_data.set_index('filename').to_csv(text_filename, sep=" ", index=False, header=False)
        # print(f'Saved data to: {text_filename}')
    except KeyError:
        # print(f'No group found for: {filename}')
    except Exception as e:
        # print(f'Error saving data for {filename}: {e}')
```

```python
filename_series = pd.Series(groupby_obj_train.groups.keys())
```

```python
filename_series.apply(save_data, args=(train_folder,groupby_obj_train))
```

```python
filename_series_test = pd.Series(groupby_obj_test.groups.keys())
filename_series_test.apply(save_data, args=(test_folder,groupby_obj_test))
```
